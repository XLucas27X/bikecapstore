```
def dfcreation(filename):
  name=spark.read.format("csv") \
          .option("header", "true") \
          .option("inferSchema", "true") \
          .option("sep",",") \
          .load("file:/databricks/driver/{}".format(filename))
  return name


brands=dfcreation('brands.csv')
categories=dfcreation('categories.csv')
customers=dfcreation('customers.csv')
order_items=dfcreation('order_items.csv')
orders=dfcreation('orders.csv')
products=dfcreation('products.csv')
staffs=dfcreation('staffs.csv')
stocks=dfcreation('stocks.csv')
stores=dfcreation('stores.csv')
shipping_companies=dfcreation('shipping_companies.csv')
order_status=dfcreation('order_status.csv')
```

```
products_df = products.join(categories, ['category_id'], 'left').join(brands, ['brand_id'], 'left')

display(products_df.take(15))
```

```
products_df = products_df.drop("brand_id", "category_id")
```

```
display(products_df.take(15))
```

```
orders_df = orders.join(order_items, ['order_id'], 'inner').join(order_status, ['status_id'], 'left').join(shipping_companies, ['company_id'], 'left')

display(orders_df.take(15))
```

```
orders_df = orders_df.drop("status_id", "company_id")
```

```
orders_df.withColumn("shipped_date", F.regexp_replace("shipped_date", "-", "")).show()
```

```
orders_df = orders_df.withColumn("shipped_date", F.regexp_replace("shipped_date", "-", ""))
orders_df = orders_df.withColumn("required_date", F.regexp_replace("required_date", "-", ""))
orders_df = orders_df.withColumn("order_date", F.regexp_replace("order_date", "-", ""))
orders_df.show()
```

```
orders_df = orders_df.withColumn('order_date', substring('order_date', 0, 8)).withColumn('required_date', substring('required_date', 0, 8))
orders_df.show()
```

```
beginDate = '2016-01-01'
endDate = '2024-12-31'
#dd-MM-yyyy

spark.sql(f"select explode(sequence(to_date('{beginDate}'), to_date('{endDate}'), interval 1 day)) as calendarDate").createOrReplaceTempView('dates')
```

```
%sql 
create temporary view  Dim_Date
as
select
  day(calendarDate)+ month(calendarDate) * 100+year(calendarDate) * 10000  as SK_date,
  calendarDate,
  year(calendarDate) AS Year,
  date_format(calendarDate, 'MMMM') as Month_Name,
  month(calendarDate) as Month,
  date_format(calendarDate, 'EEEE') as Day_Name,
  weekday(calendarDate) + 1 as DayOfWeek,
  case
    when weekday(calendarDate) < 5 then 'Y'
    else 'N'
  end as IsWeekDay,
  dayofmonth(calendarDate) as DayOfMonth,
  dayofyear(calendarDate) as DayOfYear,
  quarter(calendarDate) as QuarterOfYear
  
from
  dates
order by
  calendarDate
```

```
spark.sql("select * from Dim_Date").show()
```

```
def reverse_datekey(spark_dataframe, datekey,colno,dash=False):
    sklist=[]
    if dash==True:
        for index, i in spark_dataframe.toPandas().iterrows():
            aux=str(i[1])
            sklist.append(aux[8:]+"-"+aux[5:-3]+"-"+aux[:4])
    else:
        for index, i in spark_dataframe.toPandas().iterrows():
            aux=str(i[colno])
            sklist.append(aux[6:]+aux[4:-2]+aux[:4])  

             
    new_pandas_df = spark_dataframe.toPandas()
    new_pandas_df[datekey] = sklist

    return spark.createDataFrame(new_pandas_df)
```

```
toreverse_df=spark.sql("select * from Dim_Date")

dimDate_df=reverse_datekey(toreverse_df,"SK_date",0)

dimDate_df.show()
```

```
orders_df=reverse_datekey(orders_df,"required_date",3)
orders_df=reverse_datekey(orders_df,"order_date",2)
orders_df=reverse_datekey(orders_df,"shipped_date",4)
```
